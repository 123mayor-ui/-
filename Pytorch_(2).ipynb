```python
import pandas as pd
df_train = pd.read_csv("/content/sample_data/california_housing_train.csv")
df_test = pd.read_csv("/content/sample_data/california_housing_test.csv")

df_train
```

```python
plt.scatter(
    df_train.iloc[:, 0],
    df_train.iloc[:, 1],
    c = df_train.iloc[:, -1],
    s = 4,
    alpha = 0.7
)
plt.colorbar()
```

```python
df_train.to_numpy()
```

```python
X_train = torch.tensor(df_train.to_numpy()[:, :-1], dtype=torch.float32)
X_test = torch.tensor(df_test.to_numpy()[:, :-1], dtype=torch.float32)

y_train = torch.tensor(df_train.to_numpy()[:, -1], dtype=torch.float32)
y_test = torch.tensor(df_test.to_numpy()[:, -1], dtype=torch.float32)
```

```python
neuron_california = torch.nn.Linear(
    in_features=8,
    out_features=1
)
```

```python
neuron_california.weight
```

```python
neuron_california.weight.grad
```

```python
neuron_california.bias
```

```python
neuron_california.bias.grad
```

```python
X_train.shape
```

```python
pred = neuron_california(X_train)
pred.shape
```

```python
pred
```

```python
y_train
```

```python
loss = torch.pow(pred - y_train, 2).mean()
loss # MSE - средний квадрат ошибок
```

```python
# Для суммарного качества
loss.backward()  # Вызов расчета производных на всех параметрах модели
```

```python
neuron_california.weight.grad
```

```python
neuron_california.bias.grad
```

```python
# Обновление параметров модели на основе расчета производных
alpha = 0.0000001
# Веса w = w - alpha * dloss/dw
neuron_california.weight = torch.nn.parameter.Parameter(
neuron_california.weight - alpha * neuron_california.weight.grad
)
# Параметр b = b - alpha * dloss/db
neuron_california.bias = torch.nn.parameter.Parameter(
neuron_california.bias - alpha * neuron_california.bias.grad)
```

# Цикл обучения

```python
neuron_california = torch.nn.Linear(
    in_features=8,
    out_features=1
)

tol = 1e-4   # Критерий остановки, насколько мало должно быть изменение параметров в конце обучения
alpha = 5e-8 # Скорость обучения

for it in range(100):
    predictions = neuron_california(X_train) # Модель предсказала очередной раз
    loss = torch.pow(y_train - predictions, 2).mean() # MSE
    print(f'Квадратичная ошибка на шаге {it}:', loss)
    loss.backward()  # Производные на параметрах модели

    # w - alpha * dloss(w) / dw
    delta_w = neuron_california.weight - alpha * neuron_california.weight.grad
    neuron_california.weight = torch.nn.parameter.Parameter(
        delta_w
    )

    # b - alpha * dloss(b) / db
    delta_b = neuron_california.bias - alpha * neuron_california.bias.grad
    neuron_california.bias = torch.nn.parameter.Parameter(
        delta_b
    )

    if torch.max(delta_w) < tol and delta_b < tol:
        break
```

```python
predictions_prices = neuron_california(X_test).detach().numpy()
plt.plot(predictions_prices, predictions_prices, c='r')
plt.scatter(predictions_prices, y_test)
```
